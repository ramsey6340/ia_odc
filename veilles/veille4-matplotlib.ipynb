````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
````

Faisons un exemple:

````python
fig, ax = plt.subplots()
ax.plot([1, 5, 3, 4],[1, 2, 1, 4])
````

````python
x = np.linspace(0, 2*np.pi, 30)
y = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y)

plt.show()
````

La fonction *plot* prend 2 paramètres, le x et le y, chacune etant de type liste.<br>

````python
fig, ax = plt.subplots() # creation d'une figure avec un seul axe
````

````python
fig, ax = plt.subplots(2, 3) # creation d'une figure avec 2x3 axes
````

````python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b');
````

La signification les parametre de la methode *scatter*:<br>
```c: la color```, ```s: size```, ```data: la data```

### 3- Les interfaces explicite et implicite:
Il existe essentiellement 2 façon d'utiliser Matplotlib:<br>
- créer manuellement les figures et les axes et appellé les dessus (le style Orienté Objet)
- laisser pyplot se chargé de créer et gérer les figures et les axes et appellé les methodes dessus

Exemple de style OO (Orienté Objet)

````python
x = np.linspace(0, 2, 100)  # exemple de data.

# Notez que même dans le style OO, nous utilisons `.pyplot.figure` pour créer la figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Tracez des données sur les axes.
ax.plot(x, x**2, label='quadratic')  # Tracez plus de données sur les axes...
ax.plot(x, x**3, label='cubic')  # ... et quelques autres.
ax.set_xlabel('x label')  # Ajouter un xlabel à l'axe.
ax.set_ylabel('y label')  # Ajouter un ylabel à l'axe.
ax.set_title("Simple Plot")  # Ajouter un titre à l'axe.
ax.legend();  # Ajouter une legende.
````

Exemple de style pyplot

````python
x = np.linspace(0, 2, 100)  # exemple de data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Tracez des données sur les axes (implicites).
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();
````

````python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2);
````

````python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
````

````python
data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'});
````

````python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(np.arange(len(data1)), data1, label='data1')
ax.plot(np.arange(len(data2)), data2, label='data2')
ax.plot(np.arange(len(data3)), data3, 'd', label='data3')
ax.legend();
````

````python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# the histogram of the data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True);
````
