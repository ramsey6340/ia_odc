### 2- Creation de données<br>
Pandas comprends 2 objets principale, les **DataFrame** et les **Series**<br>
- DataFrame:<br>
Le DataFrame est un tableau. Il prend en parametère un dictionnaire python avec une clé et une valeur.<br>
Exemple:

````python
my_dict = {"Java": 2, "Python": [20, 18]}
pd.DataFrame(my_dict)
````

L'intersection de la ligne **O** et de la colonne **Java** donne **2**.<br>
L'intersection de la ligne **1** et de la colonne **Python** donne **18**.<br>
...

Les entrées de DataFrame ne sont pas limiter au entier, on peut aussi ajouté des String.<br>

````python
my_dict = {"Nom": ['Traoré', 'Sanogo', 'Sankaré'], "Prenom": ['Drissa Sidiki','Ousmane','Mamadou']}
pd.DataFrame(my_dict)
````

Nous pouvons mélanger différent type de valeur ensemble mais ça aura pas trop de sens.

````python
my_dict = {"Nom": ['Traoré', 'Sanogo', 12], "Prenom": ['Drissa Sidiki','Ousmane','Mamadou']}
pd.DataFrame(my_dict)
````

Comme on peut le voir, 12 dans le champ *Nom* n'a pas de sens.<br>
La clé du dictionnaire répresente le nom de la colonne et la valeur répresente les contenues du tableau.<br>
Par defaut, si on ne précise pas, nos indexe de ligne sont réferencés par des 0, 1, ... . Il est possible de changer ce comportement en précisant, sous forme de liste, les noms qu'on souhaite les données.<br>
Exemple:

````python
my_dict = {"Nom": ['Traoré', 'Sanogo', 'Sankaré'], "Prenom": ['Drissa Sidiki','Ousmane','Mamadou']}
pd.DataFrame(my_dict, index=['Personne 1','Personne 2','Personne 3'])
````

Nous avions dit que *pandas* est comprend 2 objets principale, parlons maintenant de la deuxième, les **Series**.<br>
- Series:<br>
Si le DataFrame est un tableau, un Series est une liste.<br>

````python
pd.Series([1,2,3,4,5])
````

Nous avons une liste de valeur de 1 à 5 avec leur indice respective de 0 à 4.<br>
Pour plus de précision, le *Series* répresente une colonne du *DataFrame*.<br>
Il est aussi possible de donner un nom aux indexes de ligne en utilisant toujour l'attribue **index**.<br>
Contrairement au *DataFrame*, le *Series* n'a pas de nom de colonne, il possède tout simplement un nom global qu'on peut préciser avec l'attribue **name**

````python
pd.Series([15, 17, 20], index=['Examen 2020','Examen 2021','Examen 2022'], name='Note Java')
````

On peut donc dire de façons bref qu'un DataFrame est tout simplement un tas de Series collé ensemble.<br>

## Indexation, Selection et Affectation
Pour acceder aux éléments d'une colonne d'un dataset, on utilise la syntaxe suivante: ```data_frame.{nom de la colonne}```.<br>
L’inconvénient avec cette technique est que si le nom de notre colonne est constitué d'au moins 2 noms séparer par des espaces, on ne peut pas utiliser cette technique.<br>
Mais il n'y a pas de soucie car pandas possède une autre technique qui est mieux adapter pour ce genre de cas, il suffit d'utiliser des crochets, et c'est ce technique qui est le plus recommandé.<br>
La syntaxe est: ```data_frame['nom de la colonne']```.<br>
Exemple:

````python
data_frame.designation
````

````python
data_frame['designation']
````

Nous obtenons ainsi un *Series* puis qu'un data frame n'est rien qu'un ensemble de Series.<br>
Pour obtenir donc seulement la valeur "Avidagos" à l'indexe 1 on fait comme suit:

````python
data_frame['designation'][1]
````

Nous avons 2 fonctions populaire qui permettent d'indexer les valeurs d'un dataset, l'un se base sur les *indexes* (la fonction```iloc```) et l'autre se base sur les labels (la fonction ```loc```).<br>

````python
data_frame.iloc[0] # on obtient la première valeur de chaque colonne
````

On peut egalement obtenir un Series en utilisant cette fonction.

````python
data_frame.iloc[:,0] # obtient tous les valeurs de la première colonne
````

On peut aussi indiquer les valeurs que l'on souhaite avoir.

````python
data_frame.iloc[1:5,[0,1]] # on obtient les valeurs de la première et de la deuxième colonne en partant de l'indexe 1 et en s'arrêtant à l'indexe 5-1 = 4
````

On peut egalement indique les indexes sous forme de liste.

````python
data_frame.iloc[[1,2,5],1] # on obtient les valeurs d'indexe 1, 2 et 3 de la deuxième colonne
````

Il est aussi possible d'indiquer des valeurs negatif pour Spécifier que l'on souhaite commencer à partant du bas vers le haut.

````python
data_frame.iloc[-5:] # on obtient les 5 dernières élément du dataset
````

Passons maintenant à la fonction *loc*.<br>
Cette fonction utilise les labels pour indexer les valeurs.<br>
Exemple:

````python
data_frame.loc[0, "country"] # obtenir la première valeur de la colonne "country"
````

````python
data_frame.loc[:,['taster_name', 'taster_twitter_handle', 'points']] # on obtient tous les entrées des colonnes 'taster_name', 'taster_twitter_handle' et 'points'
````

NB: un point important à noté comme difference entre *iloc* et *loc* est que si on a par exemple 0:50, avec *iloc* on aura les valeurs de 0 à 49 donc 50 éléments, mais avec *loc* on aura les valeurs de 0 à 50 donc 51 valeurs.

Il nous est possible de precise une colonne comme etant notre indexe, pour cela on fait comme suit:

````python
data_frame.set_index("title") # la colonne "title" devient notre indexe
````

On peut egalement poser des question à notre dataset.

````python
data_frame.country == 'Italy'
````

````python
data_frame.loc[data_frame.country == 'Italy'] # on affiche les éléments dont le champ "country" est égale à "Italy"
````

````python
data_frame.loc[(data_frame.country == 'Italy') & (data_frame.points >= 90)] # on affiche les éléments dont le champ "country" est "Italy" ET le "points" est superieur ou égale à 90
````

````python
data_frame.loc[(data_frame.country == 'Italy') | (data_frame.points >= 90)] # on affiche les éléments dont le champ "country" est "Italy" OU le "points" est superieur ou égale à 90
````

NB: Il ne faut pas oublier de mettre les logique booleen dans le parenthèses sinon tu pourra obtenir des resultats qui ne doivent pas être figurer dans le réssultat final.

Pandas incorpore des methodes qui permet directement d'obtenir des données en fonction de valeur qu'on renseigne.

````python
data_frame.loc[data_frame.country.isin(['Italy', 'France'])] # on obtient des données qui contiennent 'Italy' ou 'Frnance'
````

````python
data_frame.loc[data_frame.price.isnull()] # on obtient les données dont le champ 'price' est null
````

````python
data_frame.loc[data_frame.price.notnull()] # on obtient les données dont le champ 'price' n'est pas null 
````

### 4-Assignation de données
Pour assigner des valeurs à un DataFrame c'est très simple, il suffit de proceder comme suit:<br>
```data_frame['nom du champ'] = valeur```<br>
Exemple:

````python
data_frame['country'] = 'Mali'
data_frame['country']
````

On peut aussi ajouter d'autre champ.<br>

````python
data_frame['reverse_index'] = range(len(data_frame), 0, -1)
data_frame['reverse_index']
````
